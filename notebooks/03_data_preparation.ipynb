```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE

import joblib
```

```python
df_diabetes = pd.read_csv("../datasets/processed/diabetes_clean.csv")
df_heart = pd.read_csv("../datasets/processed/heart_clean.csv")
df_liver = pd.read_csv("../datasets/processed/liver_clean.csv")
df_kidney = pd.read_csv("../datasets/processed/kidney_clean.csv")
```

```python
def prepare_data(df, name):
    print(f"\nProcessing {name}...")

    X = df.drop("target", axis=1)
    y = df["target"]

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    joblib.dump(scaler, f"../models/{name}_scaler.pkl")

    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y,
        test_size=0.2,
        random_state=42,
        stratify=y
    )

    smote = SMOTE(random_state=42)
    X_train_sm, y_train_sm = smote.fit_resample(X_train, y_train)

    import collections
    print("Before SMOTE:", collections.Counter(y_train))
    print("After SMOTE:", collections.Counter(y_train_sm))

    np.save(f"../datasets/processed/{name}_X_train.npy", X_train_sm)
    np.save(f"../datasets/processed/{name}_X_test.npy", X_test)
    np.save(f"../datasets/processed/{name}_y_train.npy", y_train_sm)
    np.save(f"../datasets/processed/{name}_y_test.npy", y_test)

    print(f"{name} done")

    joblib.dump(list(X.columns), f"../datasets/processed/{name}_features.pkl")
```

```python
prepare_data(df_diabetes, "diabetes")
prepare_data(df_heart, "heart")
prepare_data(df_liver, "liver")
prepare_data(df_kidney, "kidney")
```
