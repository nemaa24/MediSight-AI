```python
import json
import matplotlib.pyplot as plt

datasets = ["diabetes", "heart", "liver", "kidney"]

dpd_values = []

for name in datasets:
    with open(f"../fairness/{name}_fairness.json", "r") as f:
        data = json.load(f)
        dpd_values.append(data["age_fairness"]["demographic_parity_difference"])

print("Loaded fairness data ✅")
print(dpd_values)

plt.figure()

plt.bar(datasets, dpd_values)

for i, v in enumerate(dpd_values):
    plt.text(i, v + 0.01, str(round(v, 2)), ha='center')

plt.xlabel("Datasets")
plt.ylabel("DPD (Bias)")
plt.title("Fairness Comparison Across Datasets")

plt.show()
```

```python
import json
import matplotlib.pyplot as plt

datasets = ["diabetes", "heart", "liver", "kidney"]

dpd_values = []
eod_values = []

# Load data
for name in datasets:
    with open(f"../fairness/{name}_fairness.json", "r") as f:
        data = json.load(f)
        dpd_values.append(data["age_fairness"]["demographic_parity_difference"])
        eod_values.append(data["age_fairness"]["equalized_odds_difference"])

print("Loaded fairness data ✅")
print("DPD:", dpd_values)
print("EOD:", eod_values)

# Plot
x = range(len(datasets))
width = 0.35

plt.figure(figsize=(8,5))

plt.bar([i - width/2 for i in x], dpd_values, width, label="DPD")
plt.bar([i + width/2 for i in x], eod_values, width, label="EOD")

# Add values on bars
for i, v in enumerate(dpd_values):
    plt.text(i - width/2, v + 0.01, str(round(v, 2)), ha='center')

for i, v in enumerate(eod_values):
    plt.text(i + width/2, v + 0.01, str(round(v, 2)), ha='center')

plt.xticks(x, datasets)
plt.xlabel("Datasets")
plt.ylabel("Fairness Metrics")
plt.title("Fairness Comparison (DPD vs EOD)")
plt.legend()

plt.show()
```
