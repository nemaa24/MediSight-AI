```python
import json
import matplotlib.pyplot as plt
```

```python
datasets = ["diabetes", "heart", "liver", "kidney"]

dpd_values = []

for name in datasets:
    with open(f"../fairness/{name}_fairness.json", "r") as f:
        data = json.load(f)
        dpd_values.append(data["dpd"])

print("Loaded fairness data ✅")
print(dpd_values)
```

```python
plt.figure()

plt.bar(datasets, dpd_values)

plt.xlabel("Datasets")
plt.ylabel("DPD (Bias)")
plt.title("Fairness Comparison Across Datasets")

plt.show()
```

```python
plt.figure()

plt.bar(datasets, dpd_values)

for i, v in enumerate(dpd_values):
    plt.text(i, v, str(round(v, 2)), ha='center')

plt.xlabel("Datasets")
plt.ylabel("DPD (Bias)")
plt.title("Fairness Comparison Across Datasets")

plt.show()
```
